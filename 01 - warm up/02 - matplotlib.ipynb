# Exercise 1 - Plotting your first graph

Add a cell at the end to generate a 2 D graph as follows:

x holds 1000 values between -5 and 5

z holds 1000 values between -5 and 5

y = x^2 + z^3 +6

plot x and y

name the plot(i.e.set the title) “Ploynomial_firstname” where firstname is your firstname.

Give names for the x and y axis.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
z = np.linspace(-5, 5, 1000)
y = x*2 + z*3 + 6

plt.plot(x,y)
plt.title("Ploynomial_renzo")
plt.xlabel("X (independent variable)");
plt.ylabel("y = x*2 + z*3 + 6")
plt.grid(True)
plt.show()
```

# Exercise 2 - Subplots

Add a cell at the end to generate a plot using subplot2grid with the following characteristics:

A 4 by 4 grid.

On the first row plot the function x^2 in a dashed green line.

On the second-row plot two functions, the first function x^3 in yellow color and the second function x^4 spanning three columns in red color.

On the third-row plot two functions the first X^6 in a dashed blue color and the second is X=x in magna (pink) color.

On the forth row plot one function^7 spanning all columns in dotted red.

 

Your figure should look something like this:

```python
sp1 = plt.subplot2grid((4, 4), (0,0), colspan=4);
sp1.plot(x, x**2, "g--")
sp1.axis([-1.5, 1.5, 0, 2])
sp1.yaxis.set_ticks([0, 2])
sp1.xaxis.set_ticks(np.arange(-1.5, 1.5, 0.5))

sp2 = plt.subplot2grid((4, 4), (1,0))
sp2.plot(x, x**3, "y")
sp2.axis([-1.5, 1.5, -2.5, 2.5])
sp2.xaxis.set_ticks([-1, 0, 1])

sp3 = plt.subplot2grid((4, 4), (1, 1), colspan=3)
sp3.plot(x, x**4, "r")
sp3.axis([-1.5, 1.5, 0, 5])
sp3.yaxis.set_ticks([0, 5])
sp3.xaxis.set_ticks([-1, 0, 1])

sp4 = plt.subplot2grid((4,4), (2, 0), colspan=2)
sp4.plot(x, x**6, "--b")
sp4.axis([-1.5, 1.5, 0, 10])
sp4.yaxis.set_ticks([0, 10])
sp4.xaxis.set_ticks([-1, 0, 1])

sp5 = plt.subplot2grid((4,4), (2, 2), colspan=2)
sp5.plot(x, x, "m")
sp5.axis([-1.5, 1.5, -1.5, 1.5])
sp5.yaxis.set_ticks([-1, 0, 1])
sp5.xaxis.set_ticks([-1, 0, 1])

sp6 = plt.subplot2grid((4, 4), (3,0), colspan=4)
sp6.plot(x, x**7, "r:")
sp6.axis([-1.7, 1.7, -20, 20]);
sp6.yaxis.set_ticks([-20, 0, 20])
sp6.xaxis.set_ticks(np.arange(-1.5, 1.51, 0.5))
```

# Exercise 3 - Drawing text

On the first graph showing the beautiful point add a new point name new point _firstname and display the coordinates, your figure should look something like this:

```python
x = np.linspace(-1.5, 1.5, 30)
px = 0.8
py = px**2

plt.plot(x, x**2, "b-", px, py, "ro")

plt.text(0, 1.5, "Square function\n$y = x^2$", fontsize=20, color='blue', horizontalalignment="center")
plt.text(px - 0.08, py, "Beautiful point", ha="right", weight="heavy")
plt.text(px, py, "x = %0.2f\ny = %0.2f"%(px, py), rotation=50, color='gray')

px2 = 1.3
py2 = px2**2
point_renzo = [px2, py2]

plt.plot(x, x**2, "b-", px2, py2, "ro")
plt.text(px2 - .08, py2, "point_renzo", ha="right", weight="heavy")
plt.text(px2 - .2, py2 - .33, "x2=%.2f\ny2=%.2f"%(px, py), color="gray", ha="right")

plt.show()

```

# Exercise 5 - Scatter

Add a cell to generate a scatter plot of x and y where each contains 300 numbers generated randomly between 3 and 100. Set the scale, alpha and colors as you see suitable


```python
x, y, scale = np.random.randint(3, 100, size=(3,300))

for color in ['red', 'green', 'blue']:
  plt.scatter(x, y, s=scale, c=color, edgecolors="red", alpha=.3)
  
plt.grid(True)
plt.show()
```

# Exercise 6 - Histograms

Play around with values and colors of the second graph

```python
import random

random.seed(42)

data1 = np.random.randn(400)
data2 = np.random.randn(500) - 3
data3 = np.random.randn(450) + 6
data4a = np.random.randn(200) + 3
data4b = np.random.randn(100) + 10

plt.hist(data1, bins=15, color='y', alpha=0.75, label='bar hist') # default histtype='bar'
plt.hist(data2, color='b', alpha=0.65, histtype='stepfilled', label='stepfilled hist')
plt.hist(data3, color='r', histtype='step', label='step hist')
plt.hist((data4a, data4b), color=('g','m'), alpha=0.55, histtype='barstacked', label=('barstacked a', 'barstacked b'))

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```
